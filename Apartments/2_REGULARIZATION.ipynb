```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
import scipy.stats as stats
from sklearn.preprocessing import OneHotEncoder
from sklearn import tree
from sklearn.feature_selection import mutual_info_regression
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import f_regression
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import cross_val_score
from itertools import product
from sklearn.ensemble import RandomForestRegressor
from boruta import BorutaPy as bp
import os
from pmdarima.arima import auto_arima
from sklearn.linear_model import Lasso, ElasticNet
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV, ElasticNetCV
from sklearn.metrics import mean_squared_error, r2_score

pd.set_option("display.max_columns", 500)

np.random.seed(2137) #uncomment if you want your code to be reproducible; for the purposes of our activity, let's add some randomness to the results
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')

df_train = df_train.drop(columns=['Unnamed: 0', 'log_price'])
df_test = df_test.drop(columns=['Unnamed: 0', 'log_price'])

# Separate features and target
X_train = df_train.drop(['price_z'], axis=1)
y_train = df_train['price_z']

X_test = df_test.drop(['price_z'], axis=1)
y_test = df_test['price_z']
```

```python
# log version
df_train_log = pd.read_csv('train.csv')
df_test_log = pd.read_csv('test.csv')

df_train_log = df_train_log.drop(columns=['Unnamed: 0', 'price_z'])
df_test_log = df_test_log.drop(columns=['Unnamed: 0', 'prize_z'])

# Separate features and target
X_train_log = df_train_log.drop(['log_price'], axis=1)
y_train_log = df_train_log['log_price']

X_test_log = df_test_log.drop(['log_price'], axis=1)
y_test_log = df_test_log['log_price']
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Lasso regression with cross-validation
lasso = LassoCV(cv=5, random_state=420)
lasso.fit(X_train_scaled, y_train)
```

```python
# Elastic Net regression with cross-validation
elastic = ElasticNetCV(cv=5, random_state=420)
elastic.fit(X_train_scaled, y_train)
```

```python
# Lasso predictions & evaluation
y_pred_lasso = lasso.predict(X_test_scaled)
print("Lasso RMSE:", mean_squared_error(y_test, y_pred_lasso, squared=False))
print("Lasso R2:", r2_score(y_test, y_pred_lasso))

# Elastic Net predictions & evaluation
y_pred_elastic = elastic.predict(X_test_scaled)
print("Elastic Net RMSE:", mean_squared_error(y_test, y_pred_elastic, squared=False))
print("Elastic Net R2:", r2_score(y_test, y_pred_elastic))
```
